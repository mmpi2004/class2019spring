# 현재상영영화 예매순 20위
  
  
[순위, 영화제목, 감독, 장르, 평점, 예매율] 순으로 데이터를 정리하여 csv 파일에 저장.  
-데이터 수집 출처:  네이버 영화 (https://movie.naver.com/movie/running/current.nhn?view=list&tab=normal&order=reserve)

(※ utf-8 형식의 csv파일이기 때문에 엑셀에서 열면 깨질 수 있습니다.)

```python
# _*_ coding: UTF-8 _*_
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
import re
```

```python
# 네이버 영화 / 현재상영작 / 예매순
html = urlopen("https://movie.naver.com/movie/running/current.nhn?view=list&tab=normal&order=reserve")
```

```python
bsObject = BeautifulSoup(html, "html.parser")
movies = bsObject.find_all('dl', {'class':'lst_dsc'})
```

각 영화를 구분할 수 있는 패턴을 파악하여, find_all을 통해 각 영화를 구분한 태그들의 리스트를 movies에 지정

```python
movies[0]
```

```python
with open("movie_chart.csv", "w", encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'title', 'director', 'genre', 'rating(/10)', 'ticket_sales(%)'])
    
    n=0
    for movie in movies:
        if n == 20:
            break
        n += 1
        title = movie.find(href= re.compile('/movie/bi/mi/basic.*')).get_text()
        director = movie.find('dt', {'class':'tit_t2'}).next_sibling.next_sibling.select('a')[0].get_text()
        genre = movie.find(href= re.compile("/movie/sdb/browsing/bmovie.nhn\?genre.*")).get_text()
        rating = movie.find(href= re.compile("/movie/bi/mi/point.*")).select('span.num')[0].get_text()
        ticket = movie.find_all('dt', {'class':'tit_t1'})[1].next_sibling.next_sibling.select('span')[0].get_text()
        writer.writerow([n, title, director, genre, rating, ticket])
```

영화제목 title을 찾을 때, <a href="/movie/bi/mi/basic.nhn?code=163608"\>에서 code= 뒤의 숫자만 영화마다 바뀌는 경우, re.compile('/movie/bi/mi/basic\\.nhn\?code=\d\*') 와 같이 정규표현식을 사용하여 숫자와 상관없이 매칭될 수 있도록 한다. 이 경우, title을 얻어내기 위해선 re.compile('/movie/bi/mi/basic.*') 만으로도 충분하기 때문에 후자를 사용하였다.
