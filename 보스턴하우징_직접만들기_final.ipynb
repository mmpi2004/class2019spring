```python
# descsription
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import RMSprop

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt


from tensorflow.keras.models import load_model
```

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path='boston_housing.npz')

# 훈련할 때의 loss값은 많이 줄어든 반면, test 데이터를 통해 evaluate했을 때의 loss값은 여전히 높았음. 즉, 오버피팅 문제 발생. 오버피팅을 해결하기위해서 특정 feature를 불필요하게 반영하지 않게 하기위해서 상관이 낮은 feature를 입력데이터에서 제외.
x_train = np.hstack((x_train[:, :3], x_train[:, 4:]))  # 3번째 입력(feature)인 chars를 입력데이터에서 제외. 상관이 낮다고 판단.
x_test = np.hstack((x_test[:, :3], x_test[:, 4:]))  # 3번째 입력(feature)인 chars를 입력데이터에서 제외. 상관이 낮다고 판단.

print(x_train.shape)
print(x_test.shape)
```

```python
# 실행할 때마다 같은 결과를 출력하기 위해 설정하는 부분
seed = 0
np.random.seed(seed)
tensorflow.set_random_seed(seed)

model = Sequential()
model.add(Dense(10, input_shape=[12], activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1))

model.compile(optimizer=RMSprop(lr=0.001), loss='mse',  metrics=['mae'])
model.summary()
```

```python
# train
# 데이터양이 적어서 test data로 validation data를 대신함.
hist = model.fit(x_train, y_train, epochs=3000, validation_data=(x_test, y_test))
```

```python
%matplotlib tk
fig, loss_ax = plt.subplots()

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label='test loss')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')

loss_ax.legend(loc='upper left')

plt.show()
```

```python
# test
model.evaluate(x_test, y_test)
```

```python
model.save('boston_model.h5')
```
